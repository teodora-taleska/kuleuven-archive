## if, elif and else

Let's define a list `cities` that contains 5 items:

```python
cities = ['London', 'Brussels', 'Paris', 'Toulouse', 'Leuven']
```

What happens if we ask for the 6th city, even though there are only 5?

```python
cities[5]   # Remember we start counting at zero!
```

We get an error. An `IndexError`, which basically means this index is not valid for this list. What if we would like to prevent our program from running into this error? Let's write a little program that for a given number `n`, tries to get city n, or tells us that we don't have that many cities on our list.

```python
n = 10
if n < len(cities):
    print("City", n, "is:")
    print(cities[n])
else:
    print("We don't have that many cities on our list")
```

Let's go through the code step by step. First we ask if the given value is smaller than the number of cities. The part after if evaluates to either `True` or to `False`. Let's type that in:

```python
n < len(cities)
```

Because we have 5 cities on our list, Python returns False. Let's do the same thing with a smaller value:

```python
4 < len(cities)
```

As can be expected, now the condition is `True`.

Note: since indices start counting from 0, index 4 actually refers to the 5th and last city. Therefore we use `<` meaning "strictly lower than" when comparing indices to the length of a list.

Back to our `if` statement. If the expression after `if` evaluates to `True`, our program will go on to the next line and execute the code under it. Let's try that as well:

```python
if 4 < len(cities):
    print("Index is valid!")
```

```python
if n < len(cities):
    print("Index is valid!")
```

Notice that the print statement in the last code block is not executed. That is because the value we assigned to `n` is still `10` and thus the part after `if` did not evaluate to `True`. In our little program above we used another statement besides `if`, namely `else`. It shouldn't be too hard to figure out what's going on here. The part after `else` will be executed if the condition after `if` evaluated to `False`. In English: if the value of `n` is too large, print a message stating this.

Sometimes we have several conditions that we want to treat differently. For that Python provides the `elif` statement. We use it similarly to `if` and `else`. Note however that you can only use `elif` after an `if` statement. Above we checked the size of our collection. We can also check whether a string contains certain parts, or a list contains certain elements. For example we could test whether the letter 'a' is in the word *banana*:

```python
"a" in "banana"
```

Likewise the following evaluates to `False`:

```python
"z" in "banana"
```

Let's use this in an `if-elif-else` combination:

```python
word = "subsequently"
if "a" in word:
    print(word + " contains the letter a")
elif "s" in word:
    print(word + " contains the letter s")
else:
    print("What a weird word!")
```

## Loops

Programming enables us to automate things. Once we specify how to do something once, we should also be ablo to do it many times. Loops make it possible to perform a certain action on a range of elements. For example, given a list of words, we would like to know the length of all words, not just one. Now you *could* do this by going through all the indices of a list of words and print the length of the words one at a time, taking up as many lines of code as you have indices. Needless to say, this is rather cumbersome. What's more, if we don't know the length of the list of words in advance, this strategy doesn't even work.

Python provides the `for`-statement that allows us to iterate through any iterable object and perform actions on its elements. An iterable is an object with multiple elements such as a string or a list. The basic format of a `for`-statement is:

    for x in iterable:
        # do things with x

That reads almost like English. We can print all letters of the word *banana* as follows:

```python
for letter in "banana":
    print(letter)
```

The code in the loop is executed as many times as there are letters, with a different value for the variable `letter` at each iteration. Likewise we can print all the items that are contained in a list:

```python
colors = ["yellow", "red", "green", "blue", "purple"]
for color in colors:
    print("This is the color " + color)
```

Let's create another list of cities, but this time we also include the country where the city is located. To do se, we'll create a nested list: each item is a list itself.

```python
cities_with_countries = [ ['London', 'England'], ['Brussels', 'Belgium'], ['Paris', 'France'], ['Toulouse', 'France'], ['Leuven', 'Belgium']]
```

```python
for city in cities_with_countries:
    print(city)
```

Since we know that cities in our list will always have the form `[city, country]`, we can take a shortcut and separate these two values:

```python
for city, country in cities_with_countries:
    print(city, "is in", country)
```

In the example above the variable `city` will get the first element of each item, and the variable `country` will get the second. However, if any item is not a list with exactly two elements, you will get an error.

## List Comprehensions

We can create new lists in a quick and elegant way by using a _list comprehension_. Essentially, a list comprehension loops over each item in a list, one by one, and returns something each time, and creates a new list.

For example, here is a list of strings:

`['bananas', 'apples', 'oranges', 'kiwis']`

We could use a list comprehension to loop over this list and create a new list where every item has the last letter stripped off. The resulting list would look like this:

`['banana', 'apple', 'orange', 'kiwi']`

The code for doing this is below:

```python
fruit_basket_large = ['bananas', 'apples', 'oranges', 'kiwis']
fruit_basket_small = [fruit[:-1] for fruit in fruit_basket_large]
print(fruit_basket_small)
```

The pattern is as follows:

`[return_something for each_item in list]`

The `for` and `in` are *keywords*, that is, they are special reserved words in Python. These must be present exactly in this order in every list comprehension.

The other words (`return_something`, `each_item`, `list`) are placeholders for whatever variables you are working with in your case.


> Let's look at some of the details:
 * A list comprehension goes inside square brackets (`[]`), which tells Python to create a new list.
 * `list` is the name of your list. This is usually a list that you created in a previous step.
 * The `each_item in list` part is the loop.
 * `each_item` is the name you assign to each item as it is selected by the loop. The name you choose should be something descriptive that helps you remember what it is.
 * The `return_something for` part is what happens each time it loops over an item. The `return_something` could just be the original item, or it could be something more complicated.

The most basic example is just to return each time in the loop, which will create an exact copy of the list.

Here is an example where we have taken our original list `my_list` and created a new list `new_list` with the exact same items unchanged:

```python
fruit_basket_copy = [fruit for fruit in fruit_basket_small]
print(fruit_basket_copy)
```

Why do this? There does not seem much point to creating the same list again. By adding additional operations to a list comprehension we have a powerful way to manipulate a list. We have already seen this in the `fruit` example above. Here's another example, where we loop over the `fruit_basket_small` list in order to create a new list that contains the word length of each fruit item:

```python
fruit_length = [len(fruit) for fruit in fruit_basket_small]
print(fruit_length)
```

### Filter Lists with a Condition

We can _filter_ a list by adding a _condition_ so that only certain items are included in the new list:

```python
fruit_basket_a = [fruit for fruit in fruit_basket_small if 'a' in fruit]
print(fruit_basket_a)
```

The pattern is as follows:

`[return_something for each_item in list if some_condition]`

Essentially, what we are saying here is that **if** the character "a" is **in** the item when Python loops over it, keep it and add it to the new list, otherwise ignore it and throw it away.

Thus, the new `fruit_basket_a` list has three of the four original strings in it. "banana", "apple", and "orange" have an "a" in them, while "kiwi" does not.

## Functions

Python, like most other programming languages, has functions that allow us to associate a name with a procedure. The procedure is just a series of commands that can be executed jointly by calling the function. Usually, a function takes some kind of input and computes an output (although some functions do not have any inputs or outputs). An easy example is a function that finds the area of a rectangle:

```python
def rect_area(width, height):
    return width * height

rect_area(4, 6)
```

Let’s go through this function definition bit by bit. The `def` is a special, fixed keyword that tells Python we are defining a function. Next comes the **name** of the function, which we get to choose. In this case, we chose to call it `rect_area`, but it could have been something else as well, like `area` or `rectangle_area`, as long as it does not contain spaces, nor starts with a digit, and as long as you use it consistently throughout your code.

Next come the **parameters** of the function, which are names that stand in for the inputs we will give our function later when we actually know what they are (the actual values that we are giving as input to the function are called the **arguments**). The colon tells Python that we’re ready to begin describing what our function needs to compute, and we begin the next line with four spaces.

The `return` tells Python that we want to supply the result of the following expression as the output of this function.

Note that the body of the function is indented. Python needs this space in order to tell where the function ends. You may use either tabs or spaces to create this space, as long as you are consistent. It is common practice to use four spaces by default.

After the function has been defined, we can call it to find its output given some inputs. As we can see for our example, the syntax for this is just the name of the function followed by its arguments, separated by commas.

The function is now defined, and you can call it anytime you need it (provided that you've run the previous code cell).

```python
print(rect_area(2,3.5))
```

Let's now go through our file line by line, and determine the frequency of each word. First, we'll read the file, and put all of its words in one big list. To do so, we'll make use of the **with**-statement and the **for**-loop.

We use the `with` statement to define a file stream, accessible through the variable `inFile`. We provide the `open` function with the filename on our system (`wiki_sample.txt`) and some extra parameters (`r` indicates that we want to read from the file, and `t` indicates that we want to read the file in text mode). By using the `with` statement, Python will automatically close the file for us once we're done reading from it (i.e., once the indented code block below the with statement is finished).

The `for` loop keeps executing a particular code block until a stop condition is met. In this case, we'll use it to loop through lines of the file one by one until we reach the end of the file. We write code indented, below the `for` loop, to tell Python what we want it to do with each element of the list. To each line, we apply the `strip` function, which removes trailing newline characters. Next we split the line into a list of words, by splitting on the whitespace (` `). Finally, we append the word list for each line to the `corpus` list; when we are done processing our file, the list will have accumulated all the words for our entire file. Note that we initially define `corpus` as an empty list, which we'll iteratively fill up using the `for` loop.

```python
corpus = []
with open('wiki_sample.txt', 'rt') as inFile:
    for line in inFile:
        line = line.strip()
        words = line.split(' ')
        corpus += words
```

We can now inspect the `corpus` list we have constructed; we can also check how many words are present in our list using the `len` function.

```python
print(corpus)
print('corpus length: ', len(corpus))
```

Let's now count how many times each distinct word (**type**) appears in the corpus. To count the word types, we'll make use of a **dictionary**. A dictionary is an efficient way to associate a particular *key* with a *value*. In this case, we'll use the word type as the key, and its frequency count as a value. An empty dictionary is defined using curly brackets (`{}`).

In the code below, we'll again use a `for` loop in order to traverse the entire `corpus` list, element by element (i.e. word by word). We then make use of an `if .. else` statement. This statement is used for conditional execution: when the `if`-condition is met, we execute the code block below the `if`; when the condition is not met, the `else` code block is executed instead.

In plain language, the statement reads: if the current word (key) is already present in our dictionary, then we increase the value of that word by one. If the word is not yet present, then we create the key in the dictionary, and assign a value of 1.

```python
frequencyDictionary = {}
for word in corpus:
    if word in frequencyDictionary:
        frequencyDictionary[word] += 1
    else:
        frequencyDictionary[word] = 1
```

We can now inspect the dictionary we created, and check how many distinct words there are in the corpus (i.e. check how many keys are present in the dictionary, using `len`)

```python
print(frequencyDictionary)
print(len(frequencyDictionary))
```

Let's now determine which are the most frequent words in our Wikipedia extract. To find out, we need to sort our dictionary by frequency. We can do this using the following command:

```python
sortedTypes = sorted(frequencyDictionary.items(), key=lambda x: x[1], reverse=True)
```

We use the function `items()` to get the keys and values of our dictionary as a list of tuples; the `sorted` functions sorts these tuples according to the sorting key we define (specified using the `key` argument). The `lambda` function indicates that we want to sort the elements by dictionary value (the second element) instead of the dictionary key. Finally, we set `reverse` to `True`, to indicate that we want the highest frequencies first.

We can now inspect the 10 most frequent words, together with their frequencies:

```python
print(sortedTypes[0:10])
```

Not surprisingly, words like *the* and *of* made the top 10. But what is perhaps disappointing is that so many punctuation marks made it into the top 10.

How could we get this punctuation out of our token list? First, let’s note
that strings have a method `isalpha` that returns `True` if a string only contains
alphabetical characters, and `False` otherwise. If we could somehow use this
as a criterion for whether or not we should include a token in our list, we’d
have our answer.

Python’s **list comprehension** will help us get this done with ease:

```python
filtered_corpus = [w for w in corpus if w.isalpha()]
filtered_corpus[:20]
```

There are four parts of a list comprehension. Let’s walk through them for our case. The **input sequence** `corpus` is a list with many strings inside of it. Every string in the list is bound, one by one, to the **variable `w`**. If the string’s `isalpha` method returns `True`, then the list comprehension’s **condition** is satisfied, and the **output expression** `w` is evaluated and included in the new list. If a word did not pass the `isalpha` test, it is discarded and not included in the new list.

Now, suppose we wanted to find every instance of the verb form `is` immediately followed by a regular past participle (we'll assume that a regular past participle ends with a `d`). Conceptually, in order to do this, we need to look at every item in the list and check its neighbor to the right. A list comprehension won’t easily allow us to do this, so we'll make use of a `for` loop:

```python
num_words_with_neighbor = len(corpus) - 1
positions_to_check = range(num_words_with_neighbor)
for i in positions_to_check:
  word1 = corpus[i]
  word2 = corpus[i+1]
  if word1 == 'is' and word2.endswith('d') :
    print(word1 + " " + word2)
```

Our first step is to construct a list of every index we need to check. It turns
out that this needs to be the length of the list minus one, because the last
word in the list does not have a word to the right of it. We use Python’s
`range` function to help us make this list: `range(n)` constructs a list that
contains all the numbers from `0` to `n-1`. E.g., `range(5)` produces the elements
`[0, 1, 2, 3, 4]`.

Now that we have this list, we will use the `for` loop to visit each element once. We write code indented, below the `for` loop, to tell Python what we want it to do with each element of the list. We assign the current word to the variable `word1`, and the word immediately to the right to the variable `word2`. The `if` statement then checks if the first word is equal to *is*, and if the second word ends with a *d*. If this is indeed the case, we print out the words.
